```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Titanic Data.CSV')
df.head()
```

```python
df.info()
```

```python
#.Question 1. Show the age distribution in the data using a histogram.

age_survived = df.Age[df.Survived == 1]
age_dead = df.Age[df.Survived == 0]

trace = go.Histogram(
    x=age_survived,
    opacity=0.75,
    name='Survived',
    marker=dict(color='rgba(16, 112, 2, 0.8)'),
)

trace2 = go.Histogram(
    x=age_dead,
    opacity=0.75,
    name='Deceased',
    marker=dict(color='rgba(12, 50, 196, 0.6)'),
)

data = [trace, trace2]
layout = go.Layout(barmode='group',
                   title='Age of the passengers',
                   xaxis=dict(title='Passenger Age'),
                   yaxis=dict(title='Count')
                   )

fig = go.Figure(dict(data=data, layout=layout))
fig.show()
```

```python
#.Question 2. Show the age distribution based on gender using a histogram.

gender_survived = df.Sex[df.Survived==1]
gender_deceased = df.Sex[df.Survived==0]

trace = go.Histogram(
    x=gender_survived,
    opacity=0.75,
    name='Survived',
    marker=dict(color='rgba(14, 110, 2, 0.8)'),
)

trace2 = go.Histogram(
    x=gender_deceased,
    opacity=0.75,
    name='Deceased',
    marker=dict(color='rgba(12, 50, 196, 0.6)'),
)

data = [trace, trace2]
layout = go.Layout(barmode='group',
                   title='Sex of the passengers',
                   xaxis=dict(title='Passenger Gender'),
                   yaxis=dict(title='Count')
                   )

fig = go.Figure(dict(data=data, layout=layout))
fig.show()
```

```python
import plotly.figure_factory as ff
import numpy as np
```

```python

df_corr = df.loc[:,['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
df_corr["index"] = np.arange(1,len(df_corr)+1)
```

```python

fig = ff.create_scatterplotmatrix(df_corr, diag='box', index='index', colormap='Portland',colormap_type='cat', height=700, width=700)
fig = go.Figure(fig)
fig.show()
```

```python
 # 4.Pivot the data setting the column Pclass as the columns and Fare as the values. From theresulting structure, use a boxplot to show the distribution of the values in its 3 columns.
    
pclass_fare = df.pivot(columns='Pclass', values='Fare')
pclass_fare
```

```python
trace0 = go.Box(
    y=pclass_fare[1],
    name = 'Passenger class A',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)

trace1 = go.Box(
    y=pclass_fare[2],
    name = 'Passenger class B',
    marker = dict(
        color = 'rgb(12, 128, 128)',
    )
)

trace2 = go.Box(
    y=pclass_fare[3],
    name = 'Passenger class C',
    marker = dict(
        color = 'rgb(16, 112, 2)',
    )
)

data = [trace0, trace1, trace2]
layout = go.Layout(yaxis=dict(title='Fare'))
fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
# 5. Graph the value counts of the number of passengers who survived and did not survive based on gender using a stacked bar graph.

trace1 = go.Bar(
    x=df.Sex,
    y=gender_deceased.value_counts(),
    name='Survived (gender)',
    marker=dict(color='rgba(255, 174, 255, 0.5)',
                line=dict(color='rgb(0,0,0)', width=1)),
)

trace2 = go.Bar(
    x=df.Sex,
    y=gender_survived.value_counts(),
    name='Deceased (gender)',
    marker=dict(color='rgba(255, 255, 128, 0.5)',
                line=dict(color='rgb(0,0,0)', width=1)),
)

data = [trace1, trace2]
layout = go.Layout(barmode='stack')

fig = go.Figure(dict(data=data, layout=layout))
fig.show()
```

```python
# 6. Using a scatter plot, plot the ages to the fare paid by the each passenger based on their gender.

trace1 = go.Scatter(
    x=df.Age[df.Sex == 'male'],
    y=df.Fare.value_counts(),
    mode="markers",
    name="male",
    marker=dict(color='rgba(255, 128, 255, 0.8)'),
    text=df.Fare
)

trace2 = go.Scatter(
    x=df.Age[df.Sex == 'female'],
    y=df.Fare.value_counts(),
    mode="markers",
    name="female",
    marker=dict(color='rgba(255, 128, 2, 0.8)'),
    text=df.Age
)

data = [trace1, trace2]
layout = dict(
    title='Age vs Fare based on gender',
    xaxis=dict(title='Age', ticklen=5, zeroline=False),
    yaxis=dict(title='Fare', ticklen=5, zeroline=False)
)

fig = go.Figure(dict(data=data, layout=layout))

fig.show()
```

```python
# 7. Plot a bubble plot of the ages to the fare paid by each passenger categorizing whether they survived or not. The size of each bubble should be determined by the passenger class and the name of each individual as the hover name.

pclass_size = [float(size) for size in df.Pclass]
sur_color = [float(color) for color in df.Survived]

data = [
    {
        'y': df.Fare,
        'x': df.Age,
        'mode': 'markers',
        'marker': {
            'color': sur_color,
            'size': pclass_size,
            'showscale': True
        },
        "text": df.Name
    }
]

fig = go.Figure(data)

fig.show()
```
